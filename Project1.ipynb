##Q2: Data Cleaning

Convert all string-based numerical columns (like "Price", "Odometer") into proper integers/floats.

```python
data["Price"]= pd.to_numeric(data["Price"], errors="coerce")

print(data.dtypes)
```

Standardize categorical columns (e.g., lowercase all values in make).


```python
data["Manufacturer"] = data["Manufacturer"].str.capitalize()
data["Model"] = data["Model"].str.capitalize()
data
```

Remove duplicates if any

```python

duplicates = data.duplicated(subset=["Manufacturer", "Model"])


num_duplicates = duplicates.sum()
print("Number of duplicate rows: {num_duplicates}")
```

 ## Data Analysis

Find the average car price by make **(brand)**.

```python
avg_price_by_make = data.groupby("Manufacturer")["Price"].mean().reset_index()

print(avg_price_by_make)
```

Find the most expensive brand and least expensive brand on average.

```python
most_expensive = avg_price_by_make.loc[avg_price_by_make["Price"].idxmax()]
least_expensive = avg_price_by_make.loc[avg_price_by_make["Price"].idxmin()]

print("Most expensive brand")
print(most_expensive)

print("\nLeast expensive brand:")
print(least_expensive)
```

already calculated the average price per brand, we can directly pick the max and min.


---




Group cars by fuelType and show the average price and mileage.

```python
data["Mileage"] = data["Mileage"].str.replace(" km", "", regex=True)
data["Mileage"]= pd.to_numeric(data["Mileage"], errors="coerce")
avg= data.groupby("Fuel type")[["Price", "Mileage"]].mean().reset_index()

print(avg)
```

Which transmission type (manual vs automatic) tends to cost more?


```python
avg_price_by_trans = data.groupby("Gear box type")["Price"].mean().reset_index()
print(avg_price_by_trans)
```

Use NumPy to compute:
<p>Maximum price</p>
<p>Minimum price</p>

```python
max_price = np.max(data["Price"])
min_price = np.min(data["Price"])
median_price = np.median(data["Price"])

print("Maximum Price:", max_price)
print("Minimum Price:", min_price)
print("Median Price:", median_price)
```

## Q4: Visualization


Create histograms for Price and Odometer.

```python
plt.figure(figsize=(10,5))

# Price histogram

plt.hist(data["Price"], bins=30, edgecolor="black", color="skyblue")
plt.xlabel("Price")
plt.ylabel("Number of Cars")
plt.title("Car Price Distribution")
plt.show()

# Mileage histogram

plt.hist(data["Mileage"], bins=30, edgecolor="black", color="orange")
plt.xlabel("Mileage (km)")
plt.ylabel("Number of Cars")
plt.title("Car Mileage Distribution")

plt.tight_layout()
plt.show()
```

Create a bar chart showing average car price per brand (top 10 brands only)

```python
top10 = avg_price_by_make.head(10)
plt.figure(figsize=(8,6))
plt.bar(top10["Manufacturer"], top10["Price"])

plt.title("Average Car Price per Brand")
plt.xlabel("Brand")
plt.ylabel("Average Price")
plt.xticks(rotation=45)
plt.show()
```

Create a scatter plot of Odometer vs Price.

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
plt.scatter(data["Mileage"], data["Price"], alpha=1, c="blue", edgecolor="k")

plt.title("Scatter Plot of Odometer vs Price", fontsize=16)
plt.xlabel("Odometer (Mileage in km)", fontsize=12)
plt.ylabel("Price", fontsize=12)
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()
```

Create a boxplot comparing car prices by fuelType.

```python


plt.figure(figsize=(10,6))
sns.boxplot(x="Fuel type", y="Price", data=data)

plt.title("Car Prices by Fuel Type", fontsize=16)
plt.xlabel("Fuel Type", fontsize=12)
plt.ylabel("Price", fontsize=12)
plt.show()
```

Show a pie chart of car distribution by transmission type.

```python

transmission_counts = data['Gear box type'].value_counts()


plt.figure(figsize=(6,6))
plt.pie(transmission_counts, labels=transmission_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Car Distribution by Transmission Type')
plt.show()
```
