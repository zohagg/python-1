: Data Cleaning
● Clean and convert the Price column from string to integer.
● Lowercase all values in the Make column. </p>
● Handle missing data using .fillna() and .dropna(). </p>

```python
cars = pd.read_csv("/car-sales.csv")
cars["Price"] = (
    cars["Price"]
    .str.replace("$", "", regex=False)
    .str.replace(",", "", regex=False)
    .str.replace(".00", "", regex=False)
    .astype(int)
)
cars
```

```python
cars["Make"] = cars["Make"].str.lower()
cars
```

```python
cars["Price"] = cars["Price"].fillna(cars["Price"].mean())
cars["Odometer (KM)"] = cars["Odometer (KM)"].fillna(0)
cars
```

```python
cars = cars.dropna(subset=["Price"])
cars
```

<p> Adding & Modifying Columns </p>
<p>● Add columns like Seats, Engine Size, Price_per_KM.</p>
<p>● Remove a column using .drop(). </p>

```python
cars["Seats"] = 5
cars["Engine Size"] = ["1.8L", "2.0L", "1.6L", "3.0L", "2.5L", "1.8L", "2.0L", "2.0L", "1.6L", "2.5L"]
cars["Price_per_KM"] = cars["Price"] / cars["Odometer (KM)"]
cars
```

```python
cars = cars.drop("Engine Size", axis=1)
cars
```

<p>Shuffle & Transform </p>
<p>● Shuffle rows using .sample(frac=1).</p>
<p>● Reset index using .reset_index(). </p>
<p>● Convert Odometer values to miles using .apply() and rename the column.</p>

```python
cars = cars.sample(frac=1)
```

```python
cars = cars.reset_index(drop=True)
```

```python
cars["Odometer (Miles)"] = cars["Odometer (KM)"].apply(lambda km: km * 0.621371)
```

```python
cars = cars.rename(columns={"Odometer (KM)": "Odometer_KM"})
cars
```
